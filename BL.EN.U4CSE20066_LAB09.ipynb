```python
#import statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data_given = pd.read_csv('crime_data_analysis.csv')
data_given
```

```python
target = data_given.fillna(data_given.mean())
```

```python
target
```

```python
target.describe()
```

```python
plt.scatter(x = target["unemployment"] , y = target["HDI"])
```

```python
target_temp = target[['unemployment','HDI']]
print(target_temp)
```

```python
#splitting into test and train
X = target_temp['unemployment']
y = target_temp['HDI']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
X_train = X_train.array.reshape(-1,1)
y_train = y_train.array.reshape(-1,1)
X_test = X_test.array.reshape(-1,1)
y_test = y_test.array.reshape(-1,1)
```

```python
#linear regression classifier
clf = LinearRegression()
clf.fit(X_train,y_train)
```

```python
clf_train = clf.score(X_train,y_train)
clf_test = clf.score(X_test,y_test)
print(clf_train)
print(clf_test)
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
#classification based on appliance usage from LAB 3
def explore(row: float):
  if row >= 11866.484662 :
    return 'high'
  else:
    return 'low'

average = target['Total violent crime (average)'].tolist()
classified_data = []
for i in average:
  classified_data.append(explore(i))

target['class']=classified_data
print(target)
```

```python
feature = target.iloc[:, 3:18] #for feature vector
```

```python
#splitting into test and train
from sklearn.model_selection import train_test_split
X = feature
y = target['class']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#logistic regression classifier
clf1 = LogisticRegression()
clf1.fit(X_train,y_train)
```

```python
clf1_train = clf1.score(X_train,y_train)
clf1_test = clf1.score(X_test,y_test)
print(clf1_train)
print(clf1_test)
```

```python
graph = [['Logistic regression',clf1_train,clf1_test]]
graph = pd.DataFrame(graph)
print(graph)
```

```python
plt.bar(graph[0], graph[1], label = 'Train')
plt.bar(graph[0], graph[2], label = 'Test')
plt.xlabel('Classifiers')
plt.ylabel('Accuracy Score')
plt.legend()
plt.show()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y_test)
y_test = le.fit_transform(y_test)
y_test
```

```python
clf2_train = clf2.score(X_train,y_train)
clf2_test = clf2.score(X_test,y_test)
y_pred1 = clf2.predict(X_test)
print(y_pred1)
```

```python
print(mean_squared_error(y_test, y_pred1))
```

```python
#K-NN regressor
clf9 = KNeighborsRegressor(n_neighbors = 20)
clf9.fit(X_train,y_train)
y_pred2 = clf9.predict(X_test)
mean_squared_error(y_test,y_pred2)
```
