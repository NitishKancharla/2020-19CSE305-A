```python
import pandas as pd
import numpy as np
```

```python
from sklearn import svm
```

```python
data_given = pd.read_csv('PROJECT_VC_SHEET.csv')
```

```python
data_given
```

```python
print(data_given)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

```python
data_given.describe()
```

```python
def explore(row: float):
  if row >= 24.980556:
    return 'high'
  else:
    return 'low'

Rate = data_given['Rate of Violent Crimes (2020)'].tolist()
classified_data = []
for i in Rate:
  classified_data.append(explore(i))

data_given['class']=classified_data
print(data_given)
```

```python
#scaling the data
scaler = MinMaxScaler(feature_range=(0,10))
data_given[[' Rate of Violent Crimes (2020)','Chargesheeting Rate (2020)','Total violent crime (average)','pov rate(people % below the poverty line 2021-2022)','literacy(2021)','Sex Ratio','HDI','unemployment','Police personnel per lakh of population(Actual)','MPI','Polation Density 2021 per km^2']] = scaler.fit_transform(data_given[['Rate of Violent Crimes (2020)','Chargesheeting Rate (2020)','Total violent crime (average)','pov rate(people % below the poverty line 2021-2022)','literacy(2021)','Sex Ratio','HDI','unemployment','Police personnel per lakh of population(Actual)','MPI','Polation Density 2021 per km^2']])
data = data_given
data
```

```python
#for creating a new feature vector
feature = data.iloc[:, 3:17]
```

```python
#Now we are splitting in to test and train
#splitting into test and train
X = feature
y = data_given['class']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
print(X)
```

```python
#Performing SVM classifier
clf = svm.SVC()
clf.fit(X_train, y_train)
clf.support_vectors_
print(clf.support_vectors_.shape)
```

```python
#A2. Test the accuracy of the SVM using the test set obtained from above exercise. Following code for help.
print(clf.score(X_train,y_train))
print(clf.score(X_test,y_test))
```

```python
#A3. Use the Predict function of SVC classifier to study the output values of the classifier. Relate the output value to the class value predicted. Test the accuracy of the SVM, with your own logic of class determination and comparing against the class labels, using the test set obtained from above exercise. Following code for help.
print(clf.predict(X_test))
```

```python
#A4. Experiment with various kernel functions (linear, poly, rbf & sigmoid) to train and test the accuracy of the models.
#Kernel function equal to 'poly'
#Performing SVM classifier
clf2 = svm.SVC(kernel='poly')
clf2.fit(X_train, y_train)
clf2.support_vectors_
print(clf2.support_vectors_.shape)
```

```python
print(clf2.score(X_train,y_train))
print(clf2.score(X_test,y_test))
```

```python
#Kernel function equal to 'sigmoid'
#Performing SVM classifier
clf3 = svm.SVC(kernel='sigmoid')
clf3.fit(X_train, y_train)
clf3.support_vectors_
print(clf3.support_vectors_.shape)
```

```python
print(clf3.score(X_train,y_train))
print(clf3.score(X_test,y_test))
```

```python
#Using K-NN classifier
clf4 = KNeighborsClassifier(n_neighbors = 6)
clf4.fit(X_train, y_train)
```

```python
print(clf4.score(X_train, y_train))
print(clf4.score(X_test, y_test))
```

```python
#Using MLP classifier
clf5 = MLPClassifier(solver='lbfgs', activation='logistic', hidden_layer_sizes=(2,), random_state=1)
clf5.fit(X_train, y_train)
```

```python
print(clf5.score(X_train, y_train))
print(clf5.score(X_test, y_test))
```

```python
#Kernel function equal to 'linear'
#Performing SVM classifier
clf1 = svm.SVC(kernel='linear')
clf1.fit(X_train, y_train)
clf1.support_vectors_
print(clf1.support_vectors_.shape)
```

```python
print(clf1.score(X_train,y_train))
print(clf1.score(X_test,y_test))
```
