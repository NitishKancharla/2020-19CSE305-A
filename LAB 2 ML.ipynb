```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
nltk.download('stopwords')
```

```python
bag_of_words = []

for w in unq_list:
    if w not in stop_words:
        bag_of_words.append(w)
print(bag_of_words)
print("Number of words after filtering stopwords:",len(bag_of_words))
```

```python
columns = data_given.iloc[0:36,0].to_list()
```

```python
V1 = [[] for i in range(0,37)]
for j in range(0,37):
    for i in range(0,len(bag_of_words)):
        if data_given.iloc[j,1].find(bag_of_words[i]) != -1:
            V1[j].append(1)
        else:
            V1[j].append(0)
print(V1)
print(len(V1))
```

```python
V2 = [[] for i in range(0,37)]
for j in range(0,37):
    for i in range(0, len(bag_of_words)):
        count = data_given.iloc[j,1].count(bag_of_words[i])
        V2[j].append(count)
print(V2)
```

```python
print("Length of first vector: ", len(V1))
print("Length of second vector: ", len(V2))
```

```python
#Jaccard Similarity index function
def j_set(list1, list2):
    f11 = 0
    f01 = 0
    f10 = 0
    for i in range(0,len(bag_of_words)):
        if(list1[i]==1 and list2[i]==1):
            f11 = f11+ 1
        elif(list1[i]==1 and list2[i]==0):
            f10 =f10+ 1
        elif(list1[i]==0 and list2[i]==1):
            f01 = f01+ 1
    return f11/(f01+f10+f11)
```

```python
similarity = [[] for i in range(0,37)]
for i in range(0,37):
    for j in range(0,37):
        sim = j_set(V1[i],V1[j])
        similarity[i].append(sim)
j_array = np.array(similarity)
print(j_array)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(15,10))
ax = sns.heatmap(j_array)
ax.invert_yaxis()
```

```python
from scipy import spatial
```

```python
#Cosine Similarity
cos_sim = [[] for i in range(0,37)]
for i in range(0,37):
    for j in range(0,37):
        result = 1 - spatial.distance.cosine(V2[i], V2[j])
        cos_sim[i].append(result)
plt.subplots(figsize=(15,10))
cos_arr = np.array(cos_sim)
ax = sns.heatmap(cos_arr)
ax.invert_yaxis()
```
